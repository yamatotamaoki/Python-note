```python
import matplotlib.pyplot as plt
```

```python
#リストデータのグラフを描く
data = [2,2.3,4.1,2.4,5.3,3.2,4.6]
plt.plot(data)
plt.show()
```

```python
#X軸 price Y軸 count の線グラフを描く
price = [200,300,400,500,600]
count = [31,29,25,28,26]
plt.plot(price,count)
plt.show()
```

```python
#グラフにタイトルをつける
plt.plot(price,count)
plt.title("count - price")
plt.xlabel("price")
plt.ylabel("count")
plt.show()           
```

```python
#グラフにマーカーをつける
plt.plot(price,count,marker="o")
plt.title("count - price")
plt.xlabel("price")
plt.ylabel("count")
plt.grid(True)
plt.show()
```

```python
#sinグラフを描く
import matplotlib.pyplot as plt
import math
```

```python
X = range(0,360)
Y = [math.sin(math.radians(d)) for d in X]
```

```python
plt.plot(X,Y,marker="o")
plt.ylabel("Y")
plt.xlabel("X")
plt.title("curve - graph")
plt.grid(True)
plt.show()
```

```python
#グラフ画像を保存する
plt.savefig("sin.png")
plt.show()
```

```python
## 複数のグラフを重ねる
x = range(0,360)
s = [math.sin(math.radians(d)) for d in x]
c = [math.cos(math.radians(d)) for d in x]
plt.plot(x,s)
plt.plot(x,c)
plt.savefig("sin-cos.png")
plt.show()
```

```python
#違う種類のマーカーや折れ線グラフを描く
x = [100,200,300,400,500,]
y1 = [40,65,80,100,90]
y2 = [34,56,75,91,79]
y3 = [25,47,68,76,73]
y4 = [15,40,52,64,69]
plt.plot(x,y1,marker="o",color="blue",linestyle="-")
plt.plot(x,y2,marker="v",color="red",linestyle="--")
plt.plot(x,y3,marker="^",color="green",linestyle="-.")
plt.plot(x,y4,marker="d",color="m",linestyle=":")
plt.savefig("hukusu.png")
plt.show()
```

```python
#凡例を付ける
x = [100,200,300,400,500,]
y1 = [40,65,80,100,90]
y2 = [34,56,75,91,79]
y3 = [25,47,68,76,73]
plt.plot(x,y1,marker="1",color="c",linestyle="-",label="Y1")
plt.plot(x,y2,marker="D",color="y",linestyle="-",label="Y2")
plt.plot(x,y3,marker="+",color="b",linestyle="-",label="Y3")
plt.legend(loc = "upper left")
plt.savefig("G-1")
plt.show()
```
