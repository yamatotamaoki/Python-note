```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#棒グラフ
labels = ["A","B","C","D","E","F","G","H","I","J"]
x_pos = range(0,10)
V = [91,45,17,88,47,87,49,56,67,77]
plt.bar(x_pos,V,tick_label = labels)
plt.savefig("bar.png")
plt.show()
```

```python
#横棒グラフ
labels = ["A","B","C","D","E"]
y_pos = range(0,5)
V = [91,45,17,88,47]
plt.barh(y_pos,V,tick_label = labels)
plt.savefig("barh.png")
plt.show()
```

```python
#積み上げ棒グラフ
labels = ["Green","Red","Yellow","Blue","Black","White"]
x_pos = range(0,6)
A = [34,46,54,45,56,37]
B = [17,47,55,67,38,49]
bar1 = plt.bar(x_pos,A,color="g")
bar2 = plt.bar(x_pos,B,color="c",bottom=A)
plt.xticks(x_pos,labels,rotation="vertical")
plt.legend((bar1,bar2),("man","woman"),loc = "upper right")
plt.savefig("bart.png")
plt.show()
```

```python
#散布図を描く
x = [91,45,17,88,47,87,49,56,67,23,86,20,60,67,30,41,91,55,37,14]
y = [39,17,45,32,20,14,11,48,41,13,21,40,13,14,21,40,21,18,50,25]
plt.scatter(x,y)
plt.savefig("sanpu.png")
plt.show()
```

```python
#散布図にマーカーを付ける
X1 = [79,49,24,61,37,47,70,53,48,20,2,64,77,78,78,8,6,14,62,43]
Y1 = [26,31,34,36,31,35,41,49,31,37,43,24,29,37,30,46,41,40,31,39]
X2 = [97,98,33,93,59,63,30,48,88,56,91,65,69,66,67,92,96,59,49,34]
Y2 = [62,77,60,57,46,45,49,57,60,54,53,72,46,72,59,76,67,49,42,42]
plt.scatter(X1,Y1,marker="+",color="red",label="G1")
plt.scatter(X2,Y2,marker="^",color="green",label="G2")
plt.legend(loc="upper right")
plt.savefig("sanpu2.png")
plt.show()
```

```python
#散布図のマーカーの大きさを変える
X,Y = np.random.rand(100),np.random.rand(100)
plt.scatter(X,Y,marker="o",s=500,color="cyan",alpha=0.5,linewidths=2,edgecolors="b")
plt.savefig("sanpu3.png")
plt.show()
```

```python
#第3のマーカーの大きさで表現する
X,Y = np.random.rand(100),np.random.rand(100)
V = np.random.rand(100)*1000+50
plt.scatter(X,Y,s=V,c="b",alpha=0.3,linewidths=1,edgecolors="b")
plt.savefig("sanpu4.png")
plt.show()
```

```python
#第3の値のマーカーの色で表現する
X,Y = np.random.rand(100),np.random.rand(100)
V = np.random.rand(100)
plt.scatter(X,Y,s=200,c=V,cmap="Blues",edgecolor="b")
plt.colorbar()
plt.grid(True)
plt.savefig("sanpu5.png")
plt.show()
```

```python
#円グラフ
labels=["E","D","C","B","A"]
V = [17,25,47,68,91]
ex = [0,0,0.1,0,0]
plt.pie(V,explode=ex,labels=labels,autopct='%1.1f%%',startangle=90)
plt.savefig("en.png")
plt.show()
```

```python
#グラフを左右に並べる
X1,Y1 = range(0,5),[61,45,27,88,57]
X2,Y2 = range(0,5),[17,39,46,40,27]
labels = ["A","B","C","D","E"]

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.bar(X1,Y1,color="b",tick_label=labels)
ax1.set_title("dog")

ax2 = fig.add_subplot(1,2,2)
ax2.bar(X2,Y2,color="g",tick_label=labels)
ax2.set_title("cat")
plt.savefig("dog-cat.png")
plt.show()
```

```python
#縦にグラフを並べる
X1,Y1 = range(0,7),[61,45,27,88,47,56,61]
X2,Y2 = range(0,7),[17,39,46,40,27,35,41]
labels = ["A","B","C","D","E","F","G"]
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1,facecolor="cyan")
ax1.bar(X1,Y1,color="b",tick_label=labels)
ax1.set_title("snake")

ax2 = fig.add_subplot(2,1,2,facecolor="cyan")
ax2.bar(X2,Y2,color="y",tick_label=labels)
ax2.set_title("fish")
plt.tight_layout()
plt.savefig("snake-fish.png")
plt.show()
```

```python
#上に1個、下に2個グラフを並べる
X1,Y1 = range(0,7),[61,45,27,88,47,56,61]
X2,Y2 = range(0,5),[77,49,56,47,67]
X3,Y3 = range(0,4),[56,41,67,76]
labels = ["A","B","C","D","E","F","G"]
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)
ax1.bar(X1,Y1,color="b",tick_label=labels)
ax1.set_title("dog")

ax2 = fig.add_subplot(2,2,3)
ax2.bar(X2,Y2,color="g",tick_label=labels[:5])
ax2.set_title("cat")

ax3 = fig.add_subplot(2,2,4)
ax3.bar(X3,Y3,color="c",tick_label=labels[:4])
ax3.set_title("bird")
plt.tight_layout()
plt.savefig("dog-cat-bird.png")
plt.show()
```

```python
#Y軸の最大値と最小値をax1とax2で同じにする
X1,Y1 = range(0,5),[61,45,27,88,57]
X2,Y2 = range(0,5),[17,39,46,40,27]
labels = ["A","B","C","D","E"]

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.bar(X1,Y1,color="b",tick_label=labels)
ax1.set_title("dog")
ymin,ymax = plt.ylim()

ax2 = fig.add_subplot(1,2,2)
ax2.bar(X2,Y2,color="g",tick_label=labels)
ax2.set_title("cat")
plt.ylim(ymin,ymax)
plt.savefig("dog-cat2.png")
plt.show()
```

```python
#Y軸を共有したサブプロトにグラフを描く
X1,Y1 = range(0,7),[61,45,27,88,30,79,90]
X2,Y2 = range(0,7),[45,67,85,34,22,46,73]
labels = ["aya","miki","mamiko","chie","yukiko","yuki","saori"]

fig,(ax1,ax2) = plt.subplots(nrows=1,ncols=2,sharey=True)
ax1.bar(X1,Y1,color="b",tick_label=labels)
ax1.set_title("onnnanoko")

ax2.bar(X2,Y2,color="g",tick_label=labels)
ax2.set_title("onnnanoko2")
plt.savefig("gurahu5")
plt.show()
```
